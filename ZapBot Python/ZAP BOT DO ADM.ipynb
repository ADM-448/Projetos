```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import os
import time
from urllib.parse import quote

# Configurar o serviço do ChromeDriver
servico = Service(ChromeDriverManager().install())
navegador = webdriver.Chrome(service=servico)

# Definir o caminho para o arquivo Excel e PDF
database = os.path.join('C:\\\Users\\PC\\CODIGOS\\pasta', 'telefone.xlsx')
#pdf_path = os.path.join('C:\\Users\\Usuário\\Curso_Python\\Prototipo_zapBot\\pasta', 'arquivo.pdf')

# Carregar a tabela de números de telefone
tabela = pd.read_excel(database)
lista = tabela['Telefone']

# Lista para armazenar números que não conseguiram receber a mensagem
números_falha = []

# Abrir o WhatsApp Web uma vez no início
navegador.get("https://web.whatsapp.com/")
# Aguardar o tempo necessário para o usuário escanear o QR code
WebDriverWait(navegador, 60).until(EC.presence_of_element_located((By.ID, "side")))

mensagem1 = 'Olá! Você sabia que sua startup pode conseguir recursos significativos sem precisar vender parte da empresa e sem fazer empréstimo?'
mensagem2 = 'Conseguimos te ajudar a realizar seu projeto inovador. Vamos conversar sobre como isso é possível e como sua empresa pode se beneficiar dessa oportunidade'

# Loop para enviar mensagens
for telefone in lista:
    try:
        time.sleep(5)
        # Formatar o número de telefone para o link do WhatsApp
        telefone_formatado = quote(f"+{telefone}")
        link = f"https://web.whatsapp.com/send?phone={telefone_formatado}&text={quote(mensagem1)}"
        
        navegador.get(link)
        time.sleep(10)

        # Esperar que o campo de mensagem esteja presente
        campo_mensagem = WebDriverWait(navegador, 30).until(
            EC.presence_of_element_located((By.XPATH, "//div[@contenteditable='true'][@data-tab='10']"))
        )
        time.sleep(2)
        
        # Aguardar o botão de envio estar clicável e enviar a primeira mensagem
        botao_enviar = WebDriverWait(navegador, 30).until(
            EC.element_to_be_clickable((By.XPATH, "//span[@data-icon='send']"))
        )
        botao_enviar.click()
        
        # Enviar a segunda mensagem
        time.sleep(5)
        campo_mensagem.send_keys(mensagem2)
        botao_enviar = WebDriverWait(navegador, 30).until(
            EC.element_to_be_clickable((By.XPATH, "//span[@data-icon='send']"))
        )
        botao_enviar.click()
        
        # Enviar o arquivo PDF
        '''time.sleep(15)
        clip_button = WebDriverWait(navegador, 30).until(
            EC.element_to_be_clickable((By.XPATH, "//div[@title='Anexar']"))
        )
        clip_button.click()
        
        time.sleep(2)
        attach_pdf_button = WebDriverWait(navegador, 30).until(
            EC.presence_of_element_located((By.XPATH, "//input[@accept='*']"))
        )
        attach_pdf_button.send_keys(pdf_path)'''
        
        time.sleep(15)
        send_file_button = WebDriverWait(navegador, 30).until(
            EC.element_to_be_clickable((By.XPATH, "//span[@data-icon='send']"))
        )
        send_file_button.click()
        
        # Aguardar um momento antes de enviar a próxima mensagem
        time.sleep(15)

    except Exception as e:
        print(f'Não foi possível enviar mensagem para {telefone}: {e}')
        números_falha.append(telefone)

# Fechar o navegador ao final
#navegador.quit()

# Salvar os números que falharam em um arquivo Excel
df_falha = pd.DataFrame(números_falha, columns=['Telefone'])
df_falha.to_excel('números_falha6.xlsx', index=False)

print(f"Números que não receberam a mensagem foram salvos em números_falha.xlsx")

'''
# Salvar os números que falharam em um arquivo .txt
    with open('números_falha.txt', 'w') as file:
        for numero in números_falha:
            file.write(f"{numero}\n")
        '''
```
