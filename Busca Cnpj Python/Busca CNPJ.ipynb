```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.select import Select
from selenium.webdriver.support.wait import WebDriverWait as wait
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
import pandas as pd
import pyautogui as py
import os
import time

servico = Service(ChromeDriverManager().install())
navegador = webdriver.Chrome(service=servico)

database = os.path.join('database', 'rhnumbers.xlsx')
tabela = pd.read_excel(database)

tabela['Cnpj'] = tabela['Cnpj'].str.replace('.', '', regex=False)
tabela['Cnpj'] = tabela['Cnpj'].str.replace('/', '', regex=False)
tabela['Cnpj'] = tabela['Cnpj'].str.replace('-', '', regex=False)

CNPJ = tabela['Cnpj']

rodando = True
x = 0
lista_cnpj = []
porte_tabela = []
nicho_tabela = []
zv     
for cnpjs in CNPJ:
    lista_cnpj.append(cnpjs)

def busca():
    link = "https://www.informecadastral.com.br/cnpj/" + lista_cnpj[x]
    navegador.get(link)

while rodando:
    if len(lista_cnpj) <= x:
        rodando = False
        navegador.close()
    else:
        busca()
        try:
            # Puxando o porte
            ficha_geral = WebDriverWait(navegador, 10).until(EC.presence_of_element_located((By.CLASS_NAME, 'col-lg-8')))
            ficha_port = ficha_geral.find_element(By.CLASS_NAME, 'card-body')
            busca_div = ficha_port.find_elements(By.TAG_NAME, 'div')
            porte_tabela.append(busca_div[17].find_element(By.TAG_NAME, 'p').text)
            
            try:
                # puxando o nicho
                ficha_geral = WebDriverWait(navegador, 10).until(EC.presence_of_element_located((By.CLASS_NAME, 'col-lg-8')))
                card_nicho = ficha_geral.find_elements(By.CLASS_NAME, 'card')
                nicho = card_nicho[4].find_elements(By.TAG_NAME, 'p')
                nicho_tabela.append(nicho[1].text)
            except:
                ficha_geral = WebDriverWait(navegador, 10).until(EC.presence_of_element_located((By.CLASS_NAME, 'col-lg-8')))
                card_nicho = ficha_geral.find_elements(By.CLASS_NAME, 'card')
                nicho = card_nicho[3].find_elements(By.TAG_NAME, 'p')
                nicho_tabela.append(nicho[1].text)
        except:
            nicho_tabela.append("NULL")
            porte_tabela.append("NULL")
    x += 1

tabela['Porte'] = porte_tabela
tabela['Nicho'] = nicho_tabela
tabela.to_excel(r"C:\Users\Usuário\Meus Códigos\Index-Desafio\database\Rhpronto.xlsx", sheet_name='Nome_da_Planilha', index=False)
py.press('windows')
```
